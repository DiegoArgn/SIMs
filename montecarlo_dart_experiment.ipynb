**Estimating PI in practice:**
The Dart Experiment

```python
import random
import numpy as np
import matplotlib.pyplot as plt

def monteCarloPi(numSamples):
  dartsInCircle = 0
  #Throw random darts
  for i in range(numSamples):
    x = random.uniform(0,1)
    y = random.uniform(0,1)
    #Check if darts land in circle
    if (x**2 + y**2) <= 1:
      dartsInCircle += 1
  estimate = 4 * dartsInCircle / (numSamples * 1.0)
  return estimate


  def plot_monteCarloPi(numThrows):
    # Generate logarithmic points for the number of throws
    steps = np.unique(np.logspace(1, np.log10(numThrows), num=100, dtype=int))
    estimates = []

    for step in steps:
        estimate = monteCarloPi(step)
        estimates.append(estimate)

    # Plot the results
    plt.figure(figsize=(10, 6))
    plt.plot(steps, estimates, marker='o', linestyle='-', color='b', label='Estimate of Pi')
    plt.axhline(y=np.pi, color='r', linestyle='--', label='Actual value of Pi')
    plt.xscale('log')
    plt.xlabel('Number of Throws (log scale)')
    plt.ylabel('Estimate of Pi')
    plt.title('Monte Carlo Estimation of Pi')
    plt.legend()
    plt.grid(True, which="both", linestyle="--", linewidth=0.5)
    plt.show()
```

```python
# Example usage:
print(monteCarloPi(10))
print(monteCarloPi(10000))
print(monteCarloPi(1000000))
```

```python
# Example usage
plot_monteCarloPi(1000000)
```
